Ejercicio 2 Precision y Exactitud

```python
#Importar la libreria
import numpy as np

#Dar valores de las mediciones
mediciones = np.array([9.8, 9.81, 9.79, 9.80])
#Definir el valor real
valor_real = 9.81

# Hacer las operaciones con diferentes mediciones
media = np.mean(mediciones)
error = abs(media - valor_real)

# Imprimir la mediia
print("Media:", media)
# Imprimir el error absoluto
print("Error absoluto:", error)
```

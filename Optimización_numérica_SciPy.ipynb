Ejercicio 2: Optimización numérica con SciPy

```python
#Importamos la libreria de minimizar
from scipy.optimize import minimize

#Calcular lambda
f = lambda x: (x - 2)**2

#Mostrar el resultado
resultado = minimize(f, x0=0)

#Imprimir el resultado
print(resultado.x)
```

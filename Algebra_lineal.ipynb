Ejercicio 1: Algebra lineal con Numpy

```python
#Importamos la libreria
import numpy as np

#Dar los valores de a
A = np.array([[3,2], [1,2]])

#Dar los valores de b
b = np.array([5, 5])

#Mostrar los valores de a y b, y calcular
x = np.linalg.solve(A, b)

#Imprimir la x
print(x)
```

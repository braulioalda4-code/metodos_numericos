Ejercicio 3: Visualizacion del error


```python
#Importamos la libreria numpy
import numpy as np
#Importamos el matplotlib para las graficas
import matplotlib.pyplot as plt

#Definir los valores de x para calcular
x = np.linspace(-1, 5, 100)
#Definir la y para hacer la operacion
y = (x - 2)**2


plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Funci√≥n de costo")
plt.show()
```

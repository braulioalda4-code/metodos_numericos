Ejercicio 2. Error numérico en una aproximación

Objetivo: Calcular el error al aproximar raíz 2

```python
import math  # Proporciona funciones y constantes matemáticas avanzadas para trabajar con números reales
```

```python
# Definir la raiz
real = math.sqrt(2)
# La aproximación
aprox = 1.41

# Tipo de error
error_absoluto = abs(real - aprox)
error_relativo = error_absoluto / real
# Imprimir el tipo de error
print(error_absoluto, error_relativo)
```

Ejercicio: Propagacion de error

```python
#PROPAGACION DEL ERROR
#valor de la variable
x = 2
#Error asociado
dx = 0.01

#funcion
y = x**2
#Derivada de la funcion con abs por si el valor sale negativo lo vuelva positivo
dy = abs(2*x)*dx

print("y:", y)
print("Error aproximado en y:", dy)
```
